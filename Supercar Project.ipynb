```python
#detect missing values for each column
#transform all the blank answers into missing values
df = df.replace(' ',np.nan)
ser_nan = df.isnull().sum()[df.isnull().sum()>0]
ser_nan
```

```python
df.drop(columns=['Unnamed: 13'],inplace=True)
```

```python
#fill the other columns contain missing values with mode
feature_with_nan = list(ser_nan.index.values)
feature_with_nan.remove('Unnamed: 13')


for i in feature_with_nan:
    a = df[i].value_counts().index[0]    #find the mode for certain column (only include columns have missing values)
    df[i] = df[i].fillna(a)
```

```python
#export the cleaned dataset
df.to_csv('tweets_sport-cleaned.csv',index=False)
```
